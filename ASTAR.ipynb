```python
import heapq

def Astar(streets, heuristic):
    visited = set()
    parent = {}
    start = 'ITS'
    goal = 'Tunjungan Plaza'
    fn = {start : heuristic[start]}

    pq = []
    heapq.heappush(pq, (fn[start],start))

    while pq:
        curr = heapq.heappop(pq)
        print('Cost of ',curr[1],': ',curr[0])
        
        visited.add(curr[1])
        
        if curr[1] == goal:
            temp = curr[1]
            path = ['Tunjungan Plaza']
            while temp in parent:
                path.append(parent[temp])
                temp = parent[temp]
                
                # path.reverse()
            return path[::-1]

        for s_name in streets[curr[1]]:
            if s_name not in visited:
                gn_tmp = streets[curr[1]][s_name]
                fn_tmp = gn_tmp + heuristic[s_name]
                if s_name in fn and fn_tmp < fn[s_name] or s_name not in fn:
                    fn[s_name] = fn_tmp
                    parent[s_name] = curr[1]
                    heapq.heappush(pq,(fn_tmp,s_name))
        



def main():
    heuristic = {
        'ITS' : 5950,
        'Jl Arif Rahman Hakim' : 6070,
        'Jl Basuki Rahmat' : 263,
        'Jl Kalibokor Selatan' : 3130,
        'Jl Manyar Kertoarjo' : 3640,
        'Jl Mayjen Prof Moestopo' : 2110,
        'Jl Menur Pumpungan' : 4620,
        'Jl Pemuda' : 1010,
        'Jl Raya Gubeng' : 1480,
        'Jl Raya Kertajaya Indah' : 4990,
        'Jl Ngagel' : 2490,
        'Tunjungan Plaza' : 0
    }

    distances = {
        "ITS" : {"Jl Raya Kertajaya Indah": 1500, "Jl Arif Rahman Hakim" : 2300},
        "Jl Raya Kertajaya Indah" : { "ITS" : 1500, "Jl Arif Rahman Hakim" : 1000, "Jl Manyar Kertoarjo" : 1700},
        "Jl Arif Rahman Hakim" : {"Jl Menur Pumpungan" : 1800, "ITS" : 1100, "Jl Raya Kertajaya Indah" : 1000},
        "Jl Menur Pumpungan" : {"Jl Arif Rahman Hakim" : 1800, "Jl Manyar Kertoarjo" : 2000, "Jl Kalibokor Selatan" : 2300},
        "Jl Manyar Kertoarjo" : {"Jl Mayjen Prof Moestopo" : 2400, "Jl Raya Kertajaya Indah" : 1700, "Jl Menur Pumpungan" : 2000, "Jl Raya Gubeng" : 2600},
        "Jl Mayjen Prof Moestopo" : {"Jl Manyar Kertoarjo" : 2400, "Jl Pemuda" : 1700},
        "Jl Pemuda" : {"Jl Mayjen Prof Moestopo" : 1700, "Jl Basuki Rahmat" : 500},
        "Jl Basuki Rahmat" : {"Jl Pemuda" : 500, "Jl Raya Gubeng" : 1900, "Jl Ngagel" : 2300, "Tunjungan Plaza" : 400},
        "Jl Raya Gubeng" : {"Jl Manyar Kertoarjo" : 2600, "Jl Basuki Rahmat" : 1900},
        "Jl Kalibokor Selatan" : {"Jl Menur Pumpungan" : 2300, "Jl Ngagel" : 1200},
        "Jl Ngagel" : {"Jl Basuki Rahmat" : 2300, "Jl Kalibokor Selatan" : 1200},
        "Tunjungan Plaza" : { "Jl Basuki Rahmat" : 400}
    }

    path = Astar(distances, heuristic)

    print("\nJalur Tercepat dari ITS ke Tunjungan Plaza :")
    for street in path:
        print(street)

if __name__ == '__main__':
    main()
```
