```python
from collections import deque
adj_list = {
    'Arad': {
        'Zerind': 75,
        'Timisoara': 118,
        'Sibiu': 140
    },
    'Bucharest': {
        'Giurgiu': 90,
        'Pitesti': 101,
        'Fagaras': 211,
        'Urziceni': 85
    },
    'Craiova': {
        'Dobreta': 120,
        'Rimnicu Vilcea': 146,
        'Pitesti': 138
    },
    'Dobreta': {
        'Mehadia': 75,
        'Craiova': 120
    },
    'Eforie': {
        'Hirsova': 86
    },
    'Fagaras': {
        'Sibiu': 99,
        'Bucharest': 211
    },
    'Giurgiu': {
        'Bucharest': 90
    },
    'Hirsova': {
        'Eforie': 86,
        'Urziceni': 98
    },
    'Iasi': {
        'Neamt': 87,
        'Vaslui': 92
    },
    'Lugoj': {
        'Timisoara': 111,
        'Mehadia': 70
    },
    'Mehadia': {
        'Lugoj': 70,
        'Dobreta': 75
    },
    'Neamt': {
        'Iasi': 87
    },
    'Oradea': {
        'Zerind': 71,
        'Sibiu': 151
    },
    'Pitesti': {
        'Rimnicu Vilcea': 97,
        'Craiova': 138,
        'Bucharest': 101
    },
    'Rimnicu Vilcea': {
        'Sibiu': 80,
        'Craiova': 146,
        'Pitesti': 97
    },
    'Sibiu': {
        'Oradea': 151,
        'Arad': 140,
        'Fagaras': 99,
        'Rimnicu Vilcea': 80
    },
    'Timisoara': {
        'Arad': 118,
        'Lugoj': 111
    },
    'Urziceni': {
        'Vaslui': 142,
        'Bucharest': 85,
        'Hirsova': 98
    },
    'Vaslui': {
        'Iasi': 92,
        'Urziceni': 142
    },
    'Zerind': {
        'Oradea': 71,
        'Arad': 75
    }
}


def get_path(parent: dict, src: str, dest: str) -> list:
    iterator = dest
    path = [dest]
    while iterator in parent:
        path.insert(0, parent[iterator])
        iterator = parent[iterator]
    return path


def BFS(map: dict, src: str, dest: str) -> list:
    queue = deque()
    visited = set()
    parent = dict()
    queue.append(src)
    while queue:
        current = queue.popleft()
        visited.add(current)
        if current == dest:
            return get_path(parent, src, dest)
        for city in map[current]:
            if city not in visited:
                queue.append(city)
                visited.add(city)
                parent[city] = current
    return get_path(parent, src, dest)


if __name__ == "__main__":
    print(BFS(adj_list, 'Sibiu', 'Bucharest'))
```
